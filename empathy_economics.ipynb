```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
```

## Market Externalities

### Demand-Side Empathy

```python
quantity_vals = np.arange(1, 50)
demand_vals = 80 - quantity_vals
supply_vals = 20 + quantity_vals

data = pd.DataFrame();
data["demand"] = demand_vals
data["partial_empathy_demand"] = demand_vals - 10
data["supply"] = supply_vals
data["perfect_empathy_demand"] = demand_vals - 15
```

```python
sns.lineplot(data=data["demand"], color="tab:blue", label="Normal Demand")
sns.lineplot(data=data["partial_empathy_demand"], color="tab:olive", label="Partial Empathy Demand")
sns.lineplot(data=data["perfect_empathy_demand"], color="tab:green", label="Perfect Empathy Demand")
sns.lineplot(data=data["supply"], color="tab:red", label="Supply")
plt.scatter(quantity_vals[28], demand_vals[29], color="tab:grey", s=150)
plt.scatter(quantity_vals[23] - 0.1, demand_vals[24] - 10 + 0.05, color="tab:grey", s=150)
plt.scatter(quantity_vals[21] - 0.5, demand_vals[21] - 15 - 0.5, color="tab:grey", s=150)


plt.arrow(quantity_vals[15], demand_vals[17], 0, -6, head_width=1, head_length=2, fc='black', ec='black')
plt.arrow(quantity_vals[40], demand_vals[42], 0, -6, head_width=1, head_length=2, fc='black', ec='black')
plt.annotate("Normal Equilibrium", (31.5, 49))
plt.annotate("Partial", (26, 44.2))
plt.annotate("Perfect/Optimal", (8, 41.5))
plt.xlabel("Quantity", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.title("Demand Empathy for Negative Externality", fontsize=14)
plt.xticks([])
plt.yticks([])
plt.legend()
plt.savefig("market_externality_demand.png")
plt.show()
```

### Supply-Side Empathy

```python
quantity_vals = np.arange(1, 50)
demand_vals = 80 - quantity_vals
supply_vals = 20 + quantity_vals

data = pd.DataFrame();
data["demand"] = demand_vals
data["partial_empathy_supply"] = supply_vals + 10
data["supply"] = supply_vals
data["perfect_empathy_supply"] = supply_vals + 15
```

```python
sns.lineplot(data=data["demand"], color="tab:green", label="Normal Demand")
sns.lineplot(data=data["partial_empathy_supply"], color="tab:pink", label="Partial Empathy Supply")
sns.lineplot(data=data["perfect_empathy_supply"], color="tab:purple", label="Perfect Empathy Supply")
sns.lineplot(data=data["supply"], color="tab:red", label="Supply")
plt.scatter(quantity_vals[28], demand_vals[29], color="tab:grey", s=150)
plt.scatter(quantity_vals[23] - 0.1, demand_vals[24] + 0.05, color="tab:grey", s=150)
plt.scatter(quantity_vals[21] - 0.5, demand_vals[21] - 0.5, color="tab:grey", s=150)


plt.arrow(quantity_vals[15], supply_vals[17], 0, 6, head_width=1, head_length=2, fc='black', ec='black')
plt.arrow(quantity_vals[40], supply_vals[42], 0, 6, head_width=1, head_length=2, fc='black', ec='black')
plt.annotate("Normal Equilibrium", (31.5, 49))
plt.annotate("Partial", (26, 54))
plt.annotate("Perfect/Optimal", (8, 56.7))
plt.xlabel("Quantity", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.title("Supply Empathy for Positive Externality", fontsize=14)
plt.xticks([])
plt.yticks([])
plt.legend()
plt.savefig("market_externality_supply.png")
plt.show()
```

## Bargaining

### Single Empathetic Bargainer

```python
alpha_vals = np.arange(0,1.001, 0.001)
max_acceptable_share = 1 / (alpha_vals + 1)
```

```python
sns.lineplot(x=alpha_vals, y=max_acceptable_share, color="tab:green")
plt.fill_between(alpha_vals, max_acceptable_share, color='tab:olive', alpha=0.5)
plt.fill_between(alpha_vals, max_acceptable_share, np.ones(len(alpha_vals)), color='tab:red', alpha=0.5)
sns.lineplot(x=alpha_vals, y=np.ones(len(alpha_vals)) / 2, linestyle=":", color="tab:green")

plt.annotate('50-50 Split', (0.01, 0.51), fontsize=11, alpha=0.5)
top_text = plt.text(0.65, 0.8, 'Not Nash', fontsize=20, ha='center', va='center', color='white')
top_text.set_path_effects([
    path_effects.withStroke(linewidth=2, foreground='tab:grey'),  # Black outline
    path_effects.Normal()  # Normal fill
])
bottom_text = plt.text(0.5, 0.25, 'Nash Equilibria', fontsize=20, ha='center', va='center', color='white')
bottom_text.set_path_effects([
    path_effects.withStroke(linewidth=2, foreground='tab:grey'),  # Black outline
    path_effects.Normal()  # Normal fill
])
plt.title("Acceptable Equilibria for an Empathetic Player", fontsize=14)
plt.xlabel(r"Degree of Empathy ($\alpha$)", fontsize=12)
plt.ylabel("Proportion of z recieved by player", fontsize=12)
plt.savefig("single_empathetic_bargainer.png")
plt.show()
```

### Two Empathetic Bargainers

```python
alpha_vals = np.arange(0,1.001, 0.001)
max_acceptable_share = 1 / (alpha_vals + 1)
min_acceptable_share = 1 - max_acceptable_share
```

```python
sns.lineplot(x=alpha_vals, y=max_acceptable_share, color="tab:green")
sns.lineplot(x=alpha_vals, y=min_acceptable_share, color="tab:green")
plt.fill_between(alpha_vals, min_acceptable_share, max_acceptable_share, color='tab:olive', alpha=0.5)
plt.fill_between(alpha_vals, max_acceptable_share, np.ones(len(alpha_vals)), color='tab:red', alpha=0.5)
plt.fill_between(alpha_vals, min_acceptable_share, color='tab:red', alpha=0.5)
sns.lineplot(x=alpha_vals, y=np.ones(len(alpha_vals)) / 2, linestyle=":", color="tab:green")

plt.annotate('50-50 Split', (0.01, 0.51), fontsize=11, alpha=0.5)
top_text = plt.text(0.65, 0.8, 'Not Nash', fontsize=20, ha='center', va='center', color='white')
top_text.set_path_effects([
    path_effects.withStroke(linewidth=2, foreground='tab:grey'),  # Black outline
    path_effects.Normal()  # Normal fill
])
middle_text = plt.text(0.25, 0.4, 'Nash Equilibria', fontsize=20, ha='center', va='center', color='white')
middle_text.set_path_effects([
    path_effects.withStroke(linewidth=2, foreground='tab:grey'),  # Black outline
    path_effects.Normal()  # Normal fill
])
bottom_text = plt.text(0.65, 0.2, 'Not Nash', fontsize=20, ha='center', va='center', color='white')
bottom_text.set_path_effects([
    path_effects.withStroke(linewidth=2, foreground='tab:grey'),  # Black outline
    path_effects.Normal()  # Normal fill
])
plt.title("Acceptable Equilibria for an Empathetic Player", fontsize=14)
plt.xlabel(r"Degree of Empathy ($\alpha$)", fontsize=12)
plt.ylabel("Proportion of z recieved by player", fontsize=12)
plt.savefig("two_empathetic_bargainers.png")
plt.show()
```

### Bargaining Equations

$
U_1 = x1, U_2 = x2  \\
\text{Nash Eq.: } x1 + x2 = z
$

$
U_1 = ln(x1), U_2 = ln(x2)  \\
\text{Nash Eq.: } x1 + x2 = z
$

$
U_1 = ln(x1) + \alpha * ln(x2), U_2 = ln(x2)  \\
MU_1 = \frac{1}{x_1} - \frac{\alpha}{z - x_1} \\
\frac{1}{x_1} - \frac{\alpha}{z - x_1} = 0 \Rightarrow x_1 = \frac{z}{\alpha + 1} \\
x_1 > \frac{z}{\alpha + 1} \Rightarrow x_1 < 0
$

## Prisoner's Dilemma

```python
alpha_vals = np.arange(0,1.001, 0.001)

# Base values
x4 = np.ones(len(alpha_vals))
x3 = x4 / 2
x2 = (alpha_vals + 1) / (alpha_vals + 3)
x1 = 0

# Empathetic values
em_x4 = x4 + alpha_vals * x1 
em_x3 = x3 + alpha_vals * x3
em_x2 = x2 + alpha_vals * x2
em_x1 = x1 + alpha_vals * x4
```

```python
sns.lineplot(x=alpha_vals, y=em_x4, color="tab:green", label="(Defect, Cooperate)")
sns.lineplot(x=alpha_vals, y=em_x3, color="tab:blue", label="(Cooperate, Cooperate)")
sns.lineplot(x=alpha_vals, y=em_x2, color="tab:orange", label="(Defect, Defect)")
sns.lineplot(x=alpha_vals, y=em_x1, color="tab:red", label="(Cooperate, Defect)")
plt.xlabel(r"Degree of Empathy ($\alpha$)", fontsize=12)
plt.ylabel("Player 1's Utility (Including Empathy)", fontsize=12)
plt.title("Payoffs for 'Always Existing' Empathetic Prisoner's Dilemma", fontsize=14)
plt.legend()
plt.savefig("empathetic_pd_payoffs.png")
plt.show()
```
