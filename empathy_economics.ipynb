```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Market Externalities

### Demand-Side Empathy

```python
quantity_vals = np.arange(1, 50)
demand_vals = 80 - quantity_vals
supply_vals = 20 + quantity_vals

data = pd.DataFrame();
data["demand"] = demand_vals
data["partial_empathy_demand"] = demand_vals - 10
data["supply"] = supply_vals
data["perfect_empathy_demand"] = demand_vals - 15
```

```python
sns.lineplot(data=data["demand"], color="tab:blue", label="Normal Demand")
sns.lineplot(data=data["partial_empathy_demand"], color="tab:olive", label="Partial Empathy Demand")
sns.lineplot(data=data["perfect_empathy_demand"], color="tab:green", label="Perfect Empathy Demand")
sns.lineplot(data=data["supply"], color="tab:red", label="Supply")
plt.scatter(quantity_vals[28], demand_vals[29], color="tab:grey", s=150)
plt.scatter(quantity_vals[23] - 0.1, demand_vals[24] - 10 + 0.05, color="tab:grey", s=150)
plt.scatter(quantity_vals[21] - 0.5, demand_vals[21] - 15 - 0.5, color="tab:grey", s=150)


plt.arrow(quantity_vals[15], demand_vals[17], 0, -6, head_width=1, head_length=2, fc='black', ec='black')
plt.arrow(quantity_vals[40], demand_vals[42], 0, -6, head_width=1, head_length=2, fc='black', ec='black')
plt.annotate("Normal Equilibrium", (31.5, 49))
plt.annotate("Partial", (26, 44.2))
plt.annotate("Perfect/Optimal", (8, 41.5))
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.title("Demand Empathy for Negative Externality")
plt.xticks([])
plt.yticks([])
plt.legend()
plt.savefig("market_externality_demand.png")
plt.show()
```

### Supply-Side Empathy

```python
quantity_vals = np.arange(1, 50)
demand_vals = 80 - quantity_vals
supply_vals = 20 + quantity_vals

data = pd.DataFrame();
data["demand"] = demand_vals
data["partial_empathy_supply"] = supply_vals + 10
data["supply"] = supply_vals
data["perfect_empathy_supply"] = supply_vals + 15
```

```python
sns.lineplot(data=data["demand"], color="tab:green", label="Normal Demand")
sns.lineplot(data=data["partial_empathy_supply"], color="tab:pink", label="Partial Empathy Supply")
sns.lineplot(data=data["perfect_empathy_supply"], color="tab:purple", label="Perfect Empathy Supply")
sns.lineplot(data=data["supply"], color="tab:red", label="Supply")
plt.scatter(quantity_vals[28], demand_vals[29], color="tab:grey", s=150)
plt.scatter(quantity_vals[23] - 0.1, demand_vals[24] + 0.05, color="tab:grey", s=150)
plt.scatter(quantity_vals[21] - 0.5, demand_vals[21] - 0.5, color="tab:grey", s=150)


plt.arrow(quantity_vals[15], supply_vals[17], 0, 6, head_width=1, head_length=2, fc='black', ec='black')
plt.arrow(quantity_vals[40], supply_vals[42], 0, 6, head_width=1, head_length=2, fc='black', ec='black')
plt.annotate("Normal Equilibrium", (31.5, 49))
plt.annotate("Partial", (26, 54))
plt.annotate("Perfect/Optimal", (8, 56.7))
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.title("Supply Empathy for Positive Externality")
plt.xticks([])
plt.yticks([])
plt.legend()
plt.savefig("market_externality_supply.png")
plt.show()
```
